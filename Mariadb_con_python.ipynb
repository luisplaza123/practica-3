## Consultas

```python
def obtener_instituciones(nombre_ministerio):
    cur.execute("SELECT distinct `Institución` FROM datos_modificados WHERE `Ministerio` = %s", (nombre_ministerio,))
    instituciones=[]
    for j in cur:
        instituciones.append(j)
    return instituciones
a=obtener_instituciones('Ministerio de Hacienda')
```

```python
print(obtener_instituciones('Ministerio de Hacienda'))
```

```python
def tramites_de_ministerio(ministerio):
    cur.execute("SELECT `Nombre del trámite`,`Còdigo Tràmite` FROM datos_modificados WHERE `Ministerio` = %s", (ministerio,))
    codigo=[]
    nombre=[]
    for j in cur:
        nombre.append(j[0])
        codigo.append(j[1])
    return nombre,codigo
```

```python
a,b=tramites_de_ministerio('Ministerio de Hacienda')
for j in range(0,len(a)):
    print(a[j],b[j])
```

```python
def tramites_de_institucion(institucion):
    cur.execute("SELECT `Nombre del trámite`,`Còdigo Tràmite` FROM datos_modificados WHERE `Institución` = %s", (institucion,))
    codigo=[]
    nombre=[]
    for j in cur:
        nombre.append(j[0])
        codigo.append(j[1])
    return nombre,codigo
```

```python
a,b=tramites_de_institucion('Subsecretaría de Transportes')
for j in range(0,len(a)):
    print(a[j],b[j])
```

```python
def requisitos_certificados_de_tramite(nombre):
    # Usar parámetros de consulta para evitar la inyección SQL
    cur.execute("SELECT `Detalle certificados` FROM datos_modificados WHERE `Nombre del trámite` = %s", (nombre,))
    certificados=[]
    for j in cur:
        certificados.append(j)
    return certificados
```

```python
print(requisitos_certificados_de_tramite('Acceder al concurso para el subsidio para la compra de tierras Art. 20 letra a).'))
```

```python
def requisitos_documentos_de_tramite(tramite):
    cur.execute("SELECT `Detalle documentos` FROM documentos_modificados WHERE `Nombre del trámite` = %s", (tramite,))
    documentos=[]
    for j in cur:
        documentos.append(j)
    return documentos
```

```python
print(requisitos_documentos_de_tramite('Inscripción Registro de Contratistas de Obras Mayores del Ministerio de Obras Públicas'))
```

```python
def tramites_similar(nombre_tramite):
    # Usar parámetros de consulta para evitar la inyección SQL
    nombre_para_consulta = '%' + nombre_tramite + '%'  # Cambio de nombre de la variable local
    cur.execute("SELECT `Nombre del trámite` FROM datos_modificados WHERE `Nombre del trámite` LIKE %s", (nombre_para_consulta,))
    tramites=[]
    for j in cur:
        tramites.append(j)
    return tramites
```
